```python
# not manual method

x = np.array([1,2,3,4,5,6,7,8])
y = np.array([2,5,1,7,9,3,6,10])

#Stack x with ones then transpose
A = np.vstack((x, np.ones(len(x)))).T
#rotate y to be column
Y = y[:, np.newaxis]

# direct inverse
beta = np.dot(np.dot(np.linalg.inv(np.dot(A.T, A)), A.T), Y)

# psuedo inverse
np.dot(np.linalg.pinv(A), Y)
# lstsq
np.linalg.lstsq(A, Y, rcond=None)[0]

m,b = beta[0], beta[1]

plt.title("Least Square Regression")
plt.ylabel("Y")
plt.xlabel("X")
plt.plot(x,y, "r.", label="data")
plt.plot(x, m * x + c, "b", label="least square regression")
plt.legend()
plt.show()
```
