```python
import pandas as pd


pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
# read csv from the training set
df1 = pd.read_csv("training_set_features.csv")
df2 = pd.read_csv("training_set_labels.csv")
# merge both file
df = pd.merge(df1, df2)
df.info()
```

```python
df.head(10)
```

```python
# count the number of missing values in each column
missing_values = df.isna().sum()

# show only the columns with missing values and their counts
missing_cols = missing_values[missing_values > 0]
print(missing_cols)
```

```python
cleandf = df.dropna()
cleandf.info()
```

```python
cleandf.head()
```

```python
cleandf = cleandf.drop(["employment_industry","employment_occupation", "hhs_geo_region"], axis=1)
```

```python
cleandf.head()
```

```python
cleandf['sex'].unique()
```

```python
from sklearn.preprocessing import LabelEncoder

# create an instance of label encoder
le = LabelEncoder()

# iterate over columns
for col in cleandf.columns:
    # check if column data type is object
    if cleandf[col].dtype == 'object':
        # apply label encoding
        cleandf[col] = le.fit_transform(cleandf[col].astype(str))
```

```python
# iterate over the count of columns in the DataFrame
for i in cleandf.columns:
    print(i, cleandf[i].unique())
```

```python
from sklearn.model_selection import train_test_split

X = cleandf.drop(['h1n1_vaccine', 'seasonal_vaccine'], axis=1)
y1 = cleandf["h1n1_vaccine"]
y2 = cleandf["seasonal_vaccine"]

# Split the data into training and testing sets for each target variable
X_train, X_test, y1_train, y1_test, y2_train, y2_test = train_test_split(X, y1, y2, test_size=0.3, random_state=42)

```

```python
from sklearn.linear_model import LogisticRegression

# Create the first model for y1
model1 = LogisticRegression()
model1.fit(X_train, y1_train)

# Create the second model for y2
model2 = LogisticRegression()
model2.fit(X_train, y2_train)
```

```python
y1_pred = model1.predict(X_test)
y2_pred = model2.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, f1_score

# Assuming you have trained and tested your model on X_train, y_train1, y_train2, X_test, y1_test, y2_test
# y1_pred and y2_pred are the predicted values for both targets

# Calculate accuracy and F1-score for first target variable
acc1 = accuracy_score(y1_test, y1_pred)
f1_1 = f1_score(y1_test, y1_pred)

# Calculate accuracy and F1-score for second target variable
acc2 = accuracy_score(y2_test, y2_pred)
f1_2 = f1_score(y2_test, y2_pred)

# Print the summary of performance for both targets
print("Accuracy for Target 1: ", acc1)
print("F1-score for Target 1: ", f1_1)
print("Accuracy for Target 2: ", acc2)
print("F1-score for Target 2: ", f1_2)
```

```python
train_score = model1.score(X_train, y1_train)
print(f"Training score: {train_score:.2f}")
train_score = model2.score(X_train, y2_train)
print(f"Training score: {train_score:.2f}")

# Evaluate the model on the test set
test_score = model1.score(X_test, y1_test)
print(f"Test score: {test_score:.2f}")
test_score = model2.score(X_test, y2_test)
print(f"Test score: {test_score:.2f}")
```
