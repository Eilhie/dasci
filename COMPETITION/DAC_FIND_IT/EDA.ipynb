```python
df = df.dropna(subset=['h1n1_concern','behavioral_antiviral_meds','behavioral_face_mask',
'behavioral_wash_hands','behavioral_large_gatherings','behavioral_outside_home','h1n1_knowledge','behavioral_avoidance','behavioral_touch_face'
,'household_adults','household_children'])
```

```python
# All the NULL values are populated with the mode
for col in df.columns:
    if df[col].dtype == 'object':
        df[col] = df[col].fillna(df[col].mode()[0])
    else:
        df[col] = df[col].fillna(df[col].median())
```

```python
from sklearn.preprocessing import LabelEncoder

# create an instance of label encoder
le = LabelEncoder()

# iterate over columns
for col in df.columns:
    # check if column data type is object
    if df[col].dtype == 'object':
        # apply label encoding
        df[col] = le.fit_transform(df[col].astype(str))
```

```python
df = df.drop(['race','child_under_6_months','opinion_h1n1_sick_from_vacc','opinion_seas_sick_from_vacc','household_adults','behavioral_antiviral_meds','behavioral_large_gatherings', 'behavioral_outside_home', 'behavioral_antiviral_meds','marital_status',
           'behavioral_avoidance','behavioral_face_mask','income_poverty','hhs_geo_region','employment_status','education','census_msa'],axis=1)
```

```python
df.info()
```

```python
from sklearn.model_selection import train_test_split

X = df.drop(['h1n1_vaccine', 'seasonal_vaccine'], axis=1)
y1 = df["h1n1_vaccine"]
y2 = df["seasonal_vaccine"]

# Split the data into training and testing sets for each target variable
X_train, X_test, y1_train, y1_test, y2_train, y2_test = train_test_split(X, y1, y2, test_size=0.3, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression

# Create the first model for y1
model1 = LogisticRegression()
model1.fit(X_train, y1_train)

# Create the second model for y2
model2 = LogisticRegression()
model2.fit(X_train, y2_train)
```

```python
y1_pred = model1.predict(X_test)
y2_pred = model2.predict(X_test)
```

```python
train_score = model1.score(X_train, y1_train)
print(f"Training score: {train_score:.2f}")
train_score = model2.score(X_train, y2_train)
print(f"Training score: {train_score:.2f}")

# Evaluate the model on the test set
test_score = model1.score(X_test, y1_test)
print(f"Test score: {test_score:.2f}")
test_score = model2.score(X_test, y2_test)
print(f"Test score: {test_score:.2f}")
```
