### 4. Penyelarasan kurva pertumbuhan seluruh negara dengan jumlah kasus kematian COVID-19 ≥ 25 (Bobot: 25%)

**Untuk mendapatkan gambaran perkembangan kasus covid-19 di berbagai negara, diperlukan penyelarasan kurva pertumbuhan setiap negara. Kurva hanya menampilkan informasi yang dimulai dengan hari dimana data kasus kematian COVID-19 setidaknya 25 orang.**

```python
class Negara:
    def __init__(self,data,negara):
        self.dataframe = data
        self.negara = negara
    def to_datafr(self):
        df = pd.DataFrame(self.dataframe)
        df = df.rename(columns={self.negara:'Deaths'})
        df = df.set_index([[self.negara]*len(df),df.index])
        df.index.names = ['Country','Date']
        return df
```

```python
list_negara = dateIndex_deaths.columns
list_dataframe = []
filtered_negara = []
for nama_negara in list_negara:
    filter_nama_negara = dateIndex_deaths[dateIndex_deaths[nama_negara]>=25][nama_negara]
    dataframe = Negara(filter_nama_negara,nama_negara)
    if (dataframe.to_datafr().empty):
        pass
    else:
        make_df = dataframe.to_datafr()
        list_dataframe.append(make_df)
        filtered_negara.append(nama_negara)
```

```python
growth = raw_data_confirmed.drop(columns=['Province/State','Lat','Long'])
growth = growth.groupby('Country/Region').sum()
date =[]
confirmed =[]
for i in range(len(filtered_negara)):
    for j in range(len(list_dataframe[i])):
        date.append(list_dataframe[i].index[j][1])
    for j in date:
        confirmed.append(growth.loc[filtered_negara[i],j])
#         confirmed.append(comfirmed_country_per_date.loc[j,'Total per Day'])
```

```python
more25_case_death = pd.concat(list_dataframe)
more25_case_death['Confirmed']=1
idx = 0
for i,negara in enumerate(filtered_negara):
    for j in range(len(list_dataframe[i])):
#         s.loc[(negara,date[j]),'Confirmed']=confirmed[j]
        more25_case_death.loc[negara].iloc[j,1] = confirmed[idx]
        idx+=1
more25_case_death
```

##### Tuliskan analisis anda disini:
1. Memiliki index nama negara dan tanggal dimana kasus kematian setidaknya 25 orang di negara tersebut
2. kolom Deaths berisi angka banyaknya kasus kematian
3. kolom Confirmed berisi angka banyaknya kasus positif yang terkonfirmasi

**Visualisasikan hasil diatas dan berikan/atur judul, labels, dan spesifikasi (ukuran, warna, ketebalan, dll) yang sesuai sehingga plot/kurva yang dihasilkan rapi, menarik, dan mudah dipahami.**

```python
fig,ax = plt.subplots(nrows=len(list_dataframe),figsize=(12,400))
for i,negara in enumerate(filtered_negara):
    ax[i].plot(more25_case_death.loc[negara,'Confirmed'],label='Confirmed',linewidth=3,linestyle='dashdot',color='#fcbf49')
    ax[i].set_ylabel('Confirmed Patient',fontsize=13,color='#fcbf49')
    ax[i].tick_params(axis='y',labelcolor = '#fcbf49')
    
    ax2 = ax[i].twinx()
    ax2.plot(more25_case_death.loc[negara,'Deaths'],label='Deaths Total',linewidth=3,color='#d00000')
    ax2.set_ylabel('Number of Deaths',fontsize=13,color='#d00000')
    ax2.tick_params(axis='y',labelcolor = '#d00000')
    
    lines_1,labels_1 = ax[i].get_legend_handles_labels()
    lines_2,labels_2 = ax2.get_legend_handles_labels()
    lines = lines_1 + lines_2
    labels = labels_1 + labels_2
    
    ax[i].set_title(negara,fontweight='bold',fontsize=20,fontname='Brown Std')
    ax[i].set_xticks([])
    ax[i].set_xlabel('Since {} Until {}'.format(more25_case_death.loc[negara].index[0],more25_case_death.loc[negara].index[-1]),
                    fontsize=13)
    ax[i].set_facecolor('#f8edeb')
    ax[i].legend(lines,labels,ncol=2,bbox_to_anchor=(1,1.1))

plt.show()
```

##### Tuliskan analisis anda disini:
terlihat bahwa ada kurva yang menunjukkan Kasus kematian di setiap negara yang cenderung terus meningkat. Berbeda dengan kasus kematian, kasus terkonfirmasi terlihat tidak konsisten dan terkadang tidak selaras dengan kasus kematian. Dapat saya simpulkan bahwa kemungkinan besar kasus kematian akan terus meningkat sejak munculnya kasus terkonfirmasi di negara tersebut, sehingga berapapun angka kasus terkonfirmasi nya tidak akan mempengaruhi apalagi mengurangi angka kasus kematian.
